```python
import pandas as pd
```

```python
# performance dataset
perf = pd.read_csv('otp_time_series_web.csv')
perf.head(10)
```

```python
# load factor 
rev = pd.read_csv('dom_citypairs_web.csv')
rev.head(10)
```

```python
rev.shape
```

## step 1: cap city

```python
rev['City1'] = rev['City1'].str.capitalize()
rev['City2'] = rev['City2'].str.capitalize()
rev.head(5)
```

## step 2: merge

```python
merged = pd.merge(perf, rev, 
    right_on=["City1", "City2", "Month"], 
    left_on=["Departing_Port", "Arriving_Port", "Month"])
merged.head(10)
# merged.shape
```

```python
merged.Year_y
```

## Step 3: filter out years

```python
filtered_df = merged[merged["Year_y"] >= 2021]
filtered_df.shape
```

```python
filtered_df.head(5)
```

## Step 4: drop col

```python
print(filtered_df.columns)
filtered_df.dtypes
col_drop = ['Year_x', 'Month_Num', 'City1', 'City2']
final_df = filtered_df.drop(columns=col_drop)
final_df = final_df.rename(columns={'Year_y': 'Year',
                                    'Month':'Date',
                                    'Month_num':'Month'})
```

```python
final_df.to_csv('output_file.csv', index=False)
```
